## 1. Kütüphaneleri Import Etmek

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Verisetini Çağırmak

```python
data = pd.read_csv("trends.csv")
```

## 3. Veriseti Tanımak

```python
data
```

```python
data.info()
```

```python
location = data["location"].value_counts()
year = data["year"].value_counts()
category = data["category"].value_counts()
rank = data["rank"].value_counts()
query = data["query"].value_counts()
```

```python
print("Veri kümesi {} kolon ve {} satırdan oluşuyor.".format(data.shape[1],
                                                            data.shape[0]))
print("location kolonunda {} farklı ülke var.".format(location.shape[0]))
print("year kolonunda {} farklı yıl var.".format(year.shape[0]))
print("category kolonunda {} farklı kategori var.".format(category.shape[0]))
print("rank kolonunda {} farklı sıralama var.".format(rank.shape[0]))
print("query kolonunda {} farklı sorgu var.".format(query.shape[0]))
```

## 4. Belirli çıktılar (Görselleştirmeler) Elde Etmek

```python
x = query.index
x = x[0:10]

y = query
y = y[0:10]

plt.bar(x, y)
```

```python
x = query.index
x = x[0:10]

y = query
y = y[0:10]

plt.figure(figsize=(15,8))
plt.xticks(rotation=45)
plt.bar(x, y)
```

```python
x = query.index
x = x[0:10]

y = query
y = y[0:10]

plt.figure(figsize=(15,8))
plt.rcParams["axes.facecolor"] = "black"

plt.xticks(rotation=45)
plt.title("2001 - 2019'da En Çok Trend Olan 10 Sorgu")
plt.bar(x, y, width=0.2, color="white")
plt.show()
```

```python
x = query.index
x = x[0:10]

y = query
y = y[0:10]

fig, ax = plt.subplots(1,1, figsize=(15,8))
plt.rcParams["axes.facecolor"] = "black"

plt.xticks(rotation=45)
plt.title("2001 - 2019'da En Çok Trend Olan 10 Sorgu")
plt.bar(x, y, width=0.2, color="white")
fig.text(0.55,0.65,
        
        """
            Paul Walker, Robin Williams ve Avicii
            en çok trend olan Google aramalarında 
            ilk 10'da yer alıyor.
        """, fontfamily="serif", fontsize=15, color="white")
plt.show()
```
